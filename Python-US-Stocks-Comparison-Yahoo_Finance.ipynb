# **US Stocks (^IXIC, ^GSPC, ^DJI) Comparison**<br>
Comparison of NASDAQ Composite, S&P500, and Dow Jones using Pandas, NumPy, Matplotlib, and Seaborn <br>
Data Source: https://finance.yahoo.com <br><br>
by @Anna-portfolio https://github.com/anna-portfolio

Dataframe

```python
import yfinance as yf
import numpy as np
import pandas as pd
import datetime
from functools import reduce


#get data (period of 5y)
df1 = yf.download("^IXIC", period='5y')
df2 = yf.download("^GSPC", period='5y')
df3 = yf.download("^DJI", period='5y')

#insert column with ticker into every dataframe
df1.insert(0, "Ticker", "^IXIC", True)
df2.insert(0, "Ticker", "^GSPC", True)
df3.insert(0, "Ticker", "^DJI", True)

#merge dataframes
df = pd.concat([df1, df2, df3])
print(df)
```

Data Visualization

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

Line Chart

```python
nasdaq = df.query("Ticker == '^IXIC'")
sp500 = df.query("Ticker == '^GSPC'")
dow = df.query("Ticker == '^DJI'")

nasdaq['Close'].plot(label = 'Nasdaq')
sp500['Close'].plot(label = 'S&P500')
dow['Close'].plot(label = 'Dow Jones')

plt.title('Stock Prices of NASDAQ, S&P500, and Dow Jones')
plt.legend()
plt.show()
```

Candlestick Patterns

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(x=nasdaq.index,
                open=nasdaq['Open'],
                high=nasdaq['High'],
                low=nasdaq['Low'],
                close=nasdaq['Close'])])
fig.update_layout(
    title='NASDAQ Candlestick Patterns',
    yaxis_title='NASDAQ Stock',
)
```
